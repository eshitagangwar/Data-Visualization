```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,9,10)###linspace means it will give equally space 10 nums starting from 0 to 9
```

```python
x
```

```python
x=np.linspace(0,9,6)###linspace means it will give equally space 6 nums starting from 0 to 9
```

```python
x
```

```python
y=x**2
```

```python
y
```

```python
plt.plot(x,y)#line plot is used when we use plt.plot()
plt.xlabel("x")
plt.ylabel("y")
plt.title("x-y realtionship")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("x-y relation")

plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.xlabel("x")
plt.ylabel("y")
plt.title("y-x relation")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("x-y relation")

plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.xlabel("x")
plt.ylabel("y")
plt.title("y-x relation")

plt.subplot(2,2,3)
plt.plot(x,y,'y')
plt.xlabel("x")
plt.ylabel("y")
plt.title("x-y relation")

plt.subplot(2,2,4)
plt.plot(x,y,'b')
plt.xlabel("x")
plt.ylabel("y")
plt.title("y-x relation")

plt.tight_layout()##to maintain the layout without any overlap
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,'g.-')
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,0.9,0.9])
axes2=fig.add_axes([0.9,0.9,0.4,0.5])
axes1.plot(x,y,'r.-')
axes2.plot(y,x,'g*-')
plt.show()
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y,'r.--')
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

axes[1].plot(y,x,'g*--')
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

plt.show()
```

```python
fig,axes=plt.subplots(2,2)
axes[0,0].plot(x,y,'r.--')
axes[0,0].set_xlabel("x")
axes[0,0].set_ylabel("y")

axes[0,1].plot(y,x,'g*--')
axes[0,1].set_xlabel("x")
axes[0,1].set_ylabel("y")

axes[1,0].plot(x,y,'r.--')
axes[1,0].set_xlabel("x")
axes[1,0].set_ylabel("y")

axes[1,1].plot(y,x,'g*--')
axes[1,1].set_xlabel("x")
axes[1,1].set_ylabel("y")

plt.tight_layout()
plt.show()
```

```python
fig=plt.figure()
plt.plot(x,y,'r.-',label='x-y')
plt.plot(y,x,'g-*',label='y-x')
plt.legend(loc=4)
plt.show()
```

```python
fig=plt.figure()
plt.plot(x,y,color='cyan',marker='H')
plt.show()
```
